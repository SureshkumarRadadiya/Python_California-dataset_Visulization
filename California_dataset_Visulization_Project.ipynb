#**Task 1.**
What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
dfs = pd.read_excel("housing+(1).xlsx")

# Check for missing values  Example: Dropping rows with missing median_income
# median_income = dfs.dropna(subset=["median_income"],inplace=True)
# print(median_income)

# Calculate the average median income to using the mean() method
avg_median_income = dfs['median_income'].mean()
print(f"Average Median Income: ${avg_median_income}")

# Set the style for the Seaborn plots
sns.set(style="whitegrid")

# Create a histogram using Seaborn to visualize the distribution of median income
plt.figure(figsize=(10, 6))
sns.histplot(dfs['median_income'], kde=True, color='skyblue')
plt.title('Distribution of Median Income')
plt.xlabel('Median Income')
plt.ylabel('Density')
plt.axvline(avg_median_income, color='red', linestyle='dashed', linewidth=2, label='Average Median Income')
plt.legend()
plt.show()
```

The histogram will show the distribution of median income values.

**Skewness:** histogram is skewed to the right (positively skewed), it means that there are relatively more low-income households, and the distribution has a long tail on the right side.

#**Task 6**
Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.



```python
# #Load the dataset
# dfs = pd.read_excel("housing+(1).xlsx")

def calculate_median(dfs):
    # First, sort the data
    sorted_dfs = sorted(dfs)

    # Find the length of the sorted data
    n = len(sorted_dfs )

    # Check if the number of elements is odd or even
    if n % 2 == 0:
        # If even, take the average of the middle two values
        middle1 = sorted_dfs [n // 2 - 1]
        middle2 = sorted_dfs [n // 2]
        median = (middle1 + middle2) / 2
    else:
        # If odd, take the middle value
        median = sorted_dfs [n // 2]

    return median

# Example usage:
dfs = [5, 2, 9, 1, 5, 6]  # Ex sort [1,2,5,5,6,9] 3rd and 4th location = 5+5 = 10/2 = 5
print(f"Median: {calculate_median(dfs)}")
```

#**Task 10**
Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
import pandas as pd

# Load the dataset
dfs = pd.read_csv("new_total_bedrooms.csv")

# Create a new column 'total_bedroom_size' based on the given criteria
dfs["total_bedroom_size"] = None

# Create a new column 'total_bedroom_size' using a for loop with enumerate() fuction
for j, i in enumerate(dfs["total_bedrooms"]):
   if i < 10:
     dfs.loc[j, "total_bedroom_size"] = "Small"
   elif 10 <= i <= 1000:
     dfs.loc[j, "total_bedroom_size"] = "Medium"
   else:
     dfs.loc[j, "total_bedroom_size"] = "Large"

print(dfs)
# print(dfs["total_bedroom_size"])
```

```python
import pandas as pd

#Load the dataset
dfs = pd.read_csv("new_total_bedrooms.csv")

# Create a new column 'total_bedroom_size' based on the given criteria with pd.cut() function
dfs['total_bedroom_size'] = pd.cut(dfs['total_bedrooms'],
                                    bins=[0, 10, 1000, float('inf')],
                                    labels=['Small', 'Medium', 'Large'],
                                    right=False, include_lowest=True)

# 'total_bedroom_size' is now a new column with values 'Small', 'Medium', or 'Large'.
print(dfs.head())
# print(df['total_bedroom_size'])
```

```python
import pandas as pd

# Load  dataset
dfs = pd.read_csv("new_total_bedrooms.csv")

# Create a new column 'total_bedroom_size' using a for loop with append() fuction
total_bedroom_size = []

for total_bedrooms in dfs['total_bedrooms']:
    if total_bedrooms <= 10:
        total_bedroom_size.append("Small")
    elif 10 < total_bedrooms < 1000:
        total_bedroom_size.append("Medium")
    else:
        total_bedroom_size.append("Large")

dfs['total_bedroom_size'] = total_bedroom_size

# 'total_bedroom_size' is now a new column with values 'Small', 'Medium', or 'Large'.
print(dfs.head())
```
